```python
# df[(df['wS']/df['wS'].mean()) >= 1.25]['wS']
# N
import math
distanceN = 0
for i in range(1, len(df)):
    distanceN += math.sqrt((df['xN'][i] - df['xN'][i - 1])**2 + (df['yN'][i] - df['yN'][i - 1])**2)
print(((distanceN / 10) * 30) / 100, 'm')
```
